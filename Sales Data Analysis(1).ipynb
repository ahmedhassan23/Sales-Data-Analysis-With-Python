## EXPLORATORY DATA ANALYSIS


### * WHAT IS THE OVERALL SALES TREND?


```python
data['order_date'].min()
```

```python
data['order_date'].max()
```

```python
# Getting month year from order_date
data['month_year']=data['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
data['month_year'].values
```

```python
data.groupby('month_year').sum()
```

```python
# grouping month_year by sales
trend_sales=data.groupby('month_year').sum()['sales'].reset_index()
trend_sales.head(10)
```

```python
plt.figure(figsize=(16,5))
plt.plot(trend_sales['month_year'],trend_sales['sales'])
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
product_sales=data.groupby('product_name').sum()['sales'].reset_index()
product_sales[:10]
```

#### TOP 10 PRODUCTS BY SALES

```python
product_sales.sort_values(by=['sales'],inplace=True, ascending=False)
product_sales[:10]
```

#### THE MOST SELLING PRODUCTS

```python
product_quantity=pd.DataFrame(data.groupby('product_name').sum()['quantity'])
```

```python
product_quantity.sort_values(by=['quantity'],inplace=True,ascending=False)
product_quantity[:10]
```

### THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY

```python
cat_subcat=pd.DataFrame(data.groupby(['category', 'sub_category']).sum()['profit'])
cat_subcat
```

```python
# Sorting the values
cat_subcat.sort_values(['category','profit'],inplace=True ,ascending=False)
```

```python
cat_subcat
```
